```python
import shutil
import os
```

```python
imagesPath = './splitted_data/images/'
dotsPath = './splitted_data/dots/'
masksPath = './splitted_data/masks/'
coordinateTextsPath = './splitted_data/coordinate_texts/'

imageSetsTrainPath = './splitted_data/image_sets/training.txt'
imageSetsTestPath = './splitted_data/image_sets/test.txt'
imageSetsValidationPath = './splitted_data/image_sets/validation.txt'
imageSetsTrainValPath = './splitted_data/image_sets/trainval.txt'

dataTrainImagesPath = './splitted_data/train/images/'
dataTrainDotsPath = './splitted_data/train/dots/'
dataTrainMasksPath = './splitted_data/train/masks/'
dataTrainCoordinateTextsPath = './splitted_data/train/coordinate_texts/'

dataTestImagesPath = './splitted_data/test/images/'
dataTestDotsPath = './splitted_data/test/dots/'
dataTestMasksPath = './splitted_data/test/masks/'
dataTestCoordinateTextsPath = './splitted_data/test/coordinate_texts/'

dataValidationImagesPath = './splitted_data/validation/images/'
dataValidationDotsPath = './splitted_data/validation/dots/'
dataValidationMasksPath = './splitted_data/validation/masks/'
dataValidationCoordinateTextsPath = './splitted_data/validation/coordinate_texts/'

dataTrainValImagesPath = './splitted_data/train_val/images/'
dataTrainValDotsPath = './splitted_data/train_val/dots/'
dataTrainValMasksPath = './splitted_data/train_val/masks/'
dataTrainValCoordinateTextsPath = './splitted_data/train_val/coordinate_texts/'
```

```python
def splittingData(n_samples=None):
    image_sets_path_list = {
        'train': imageSetsTrainPath,
        'test': imageSetsTestPath,
        'validation': imageSetsValidationPath,
        'trainval': imageSetsTrainValPath
    }
    for key in image_sets_path_list:
        with open(image_sets_path_list[key]) as file:
            for line in file:
                file_name = os.path.splitext(line.rstrip())[0]
                source_image_path = imagesPath + file_name + '.jpg'
                source_coordinate_text_path = coordinateTextsPath + file_name + '.txt'
                source_dot_path = dotsPath + file_name + 'dots.png'
                source_mask_path = masksPath + file_name + 'mask.mat'

                if key == 'train':
                    shutil.copy2(source_image_path, dataTrainImagesPath)
                    shutil.copy2(source_coordinate_text_path, dataTrainCoordinateTextsPath)
                    shutil.copy2(source_dot_path, dataTrainDotsPath)
                    shutil.copy2(source_mask_path, dataTrainMasksPath)
                    pass
                if key == 'test':
                    shutil.copy2(source_image_path, dataTestImagesPath)
                    shutil.copy2(source_coordinate_text_path, dataTestCoordinateTextsPath)
                    shutil.copy2(source_dot_path, dataTestDotsPath)
                    shutil.copy2(source_mask_path, dataTestMasksPath)
                    pass
                if key == 'validation':
                    shutil.copy2(source_image_path, dataValidationImagesPath)
                    shutil.copy2(source_coordinate_text_path, dataValidationCoordinateTextsPath)
                    shutil.copy2(source_dot_path, dataValidationDotsPath)
                    shutil.copy2(source_mask_path, dataValidationMasksPath)
                    pass
                if key == 'trainval':
                    shutil.copy2(source_image_path, dataTrainValImagesPath)
                    shutil.copy2(source_coordinate_text_path, dataTrainValCoordinateTextsPath)
                    shutil.copy2(source_dot_path, dataTrainValDotsPath)
                    shutil.copy2(source_mask_path, dataTrainValMasksPath)
                    pass
```

```python
splittingData()
```
